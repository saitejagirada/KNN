```python
import numpy as np
import torch
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from collections import Counter
```

```python
def pred(a,b):
    accuracy=torch.sum(a==b).item()/len(a)
    return accuracy*100
```
